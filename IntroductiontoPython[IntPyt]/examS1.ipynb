## Exercice 1: functions (2 points + 0.5 bonus points) <a name="1"></a>

1. **(1 pt)** Create a function `add(a, b = 1)` that returns the value $a + 2\times b$ given two numbers `a` and `b`.

**Example:**
```
>>> add(6, 2)
10
>>> add(10, -1)
8
>>> add(10)

```

2. **(1 pt)** Create a function `substract(a, b = 1)` that returns the value $a - 2\times b$ given two numbers `a` and `b`.

**Example:**
```
>>> substract(6, 2)
2
>>> substract(10, -1)
12
>>> substract(10)

```

3. (<font color = "red"><b>Bonus point</b></font>) With the functions `add` and `substract`, create a function `even(a)` that returns 
    - $a + 4$ if $a$ is an even number
    - $a - 4$ if $a$ is an odd number
    - $0$ if `a` is not an integer

**Example:**
```
>>> even(10)
12
>>> even(3)
1
>>> even(3.)
0
```

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE THE FUNCTION: add
## question 1
def add(a, b = 1):
    return a + 2 * b
````

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE THE FUNCTION: substract
## question 2
def substract(a, b = 1):
    return a - 2 * b
````

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE THE FUNCTION: even
## question 3 (BONUS)
def even(a):
    """
    due to the issue on question 3, returns are a + 2 and a - 2
    """
    if a % 2 == 0:
        return add(a, 1)
    elif a % 2 == 1:
        return substract(a, 1)
    else:
        return 0
````

1. **(1 pt)** Create a function `vandermonde(n)` which generate the following $n\times n$ array (you can use explicit loops):

$$
V_n = 
\begin{pmatrix}
1^0 & 1^1 & \cdots & 1^n \\
2^0 & 2^1 & \cdots & 2^n \\
\vdots&^  \vdots  & \ddots & \vdots  \\
n^0 & n^1 & \cdots & n^n
\end{pmatrix}
$$
And name `arr3` the array generated by `vandermonde(6)`

**Example:**
```
>>> vandermonde(3)
array([[1, 1, 1],
       [1, 2, 4],
       [1, 3, 9]])
>>> vandermonde(4)
array([[ 1,  1,  1,  1],
       [ 1,  2,  4,  8],
       [ 1,  3,  9, 27],
       [ 1,  4, 16, 64]])
>>> arr3 = vandermonde(6)
```

2. **(1 pt)** Replace all the odd values of `arr3` by -1.

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE: vandermonde and arr3
## question 1
def vandermonde(n):
    vlist = []
    for i in range(1, n + 1):
        vlist.append([])
        for j in range(n):
            vlist[i - 1].append(i ** j)
    return vlist

arr3 = vandermonde(6)
````

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD UDPATE THE ARRAY: arr3
## question 2
for i in range(len(arr3)):
    for j in range(len(arr3[i])):
        arr3[i][j] = -1 if arr3[i][j] % 2 == 1 else arr3[i][j]
````
